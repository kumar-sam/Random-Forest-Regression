```python
# Assignment 24
"""
Question:
In this assignment students will build the random forest model after normalizing the
variable to house pricing from boston data set.
"""
#Code to loading data and modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
boston = datasets.load_boston()
features_df = pd.DataFrame(boston.data, columns=boston.feature_names)
targets_df = boston.target
```

```python
# dividing data into test train sets
x_train,x_test,y_train,y_test = train_test_split(features_df,targets_df,test_size=.2,random_state=42)
print('x_train: ',x_train.shape)
print('x_test: ',x_test.shape)
print('y_train: ',y_train.shape)
print('y_test: ',y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor # importing RandomForestRegressor class from sklearn
from sklearn.metrics import mean_squared_error # importing mean_squared_error from sklearn
min_error=1000000;
for i in range(10,50,5): # nested for loops check for perfect value of params like n_estimators, max_depth and max_features
    for j in range(1,10):
        for k in range(4,10):
            regressor=RandomForestRegressor(n_estimators=i,max_depth=j,max_features=k,random_state=42) # making regressor object
            regressor.fit(x_train,y_train) # training the model
            y_pred=regressor.predict(x_test) # predicting the value on test data
            error=mean_squared_error(y_test,y_pred) # mean squared error calculation
            if(error<min_error): # if is used to find optimum min_error
                min_error=error
                error_i=i
                error_j=j
                error_k=k
            print(' n_estimators= ',i,' max_depth= ',j,' max_features= ',k,'    error= ',error)
print('\n\n\n n_estimators= ',error_i,' max_depth= ',error_j,' max_features= ',error_k,'    error= ',min_error)
```

 hence we get minimum test error when n_estimators=  15  max_depth=  8  max_features=  7     error=  5.999488651566085

```python
# hence we will make our new model with n_estimators=  15  max_depth=  8  max_features=  7
from sklearn.metrics import mean_absolute_error
regressor=RandomForestRegressor(n_estimators=15,max_depth=8,max_features=7,random_state=42) # making regressor object
regressor.fit(x_train,y_train) # training the model
y_pred=regressor.predict(x_test) # predicting the value on test data
mse_error=mean_squared_error(y_test,y_pred) # mean squared error calculation
mae_error=mean_absolute_error(y_test,y_pred) # mean absolute error calculation
print("Mean squared error:",mse_error)
print("Mean absolute error:",mae_error)
```
